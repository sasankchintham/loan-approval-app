```python
data_numeric = df.select_dtypes(include='number')
cols_to_cap = [col for col in data_numeric.columns if col != 'loan_status']

for col in cols_to_cap:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Protect realistic mins
    if col == 'person_age':
        lower = max(lower, 18)
    if col == 'person_emp_exp':
        lower = max(lower, 0)
    
    df[col] = df[col].clip(lower=lower, upper=upper)

print("Outliers capped. loan_status preserved.")
```

```python
import pandas as pd

X = data
corr_matrix = X.corr().abs()

# find highly correlated pairs
high_corr_pairs = [
    (i, j, corr_matrix.loc[i, j])
    for i in corr_matrix.columns
    for j in corr_matrix.columns
    if i != j and corr_matrix.loc[i, j] > 0.9
]

pd.DataFrame(high_corr_pairs, columns=["Feature_1", "Feature_2", "Correlation"])
```

```python
from sklearn.model_selection import train_test_split
X=data.drop('loan_status', axis=1)
y=data['loan_status']

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=44,test_size=0.2,stratify=y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Best Parameters: {'class_weight': 'balanced', 'criterion': 'entropy', 'max_depth': 6, 'min_samples_leaf': 20, 'min_samples_split': 2}
```

```python
dtc=DecisionTreeClassifier(class_weight='balanced',
                           criterion='entropy',max_depth=6,
                           min_samples_leaf=20,min_samples_split=2,random_state=42)
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred=dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,f1_score,precision_score,confusion_matrix

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = dtc.predict(X_test)

print("ðŸŽ¯ Accuracy:", accuracy_score(y_test, y_pred))
print("\nðŸ“Š Classification Report:\n", classification_report(y_test, y_pred))
print("\nðŸŒ€ Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\n precision:",precision_score(y_test,y_pred))
print("\n f1_score:\n",f1_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'criterion': ['gini', 'entropy'],        # measure of purity
    'max_depth': [4, 6, 8, 10, 12, 15, None],
    'min_samples_split': [2, 5, 10, 20, 50],
    'min_samples_leaf': [1, 3, 5, 10, 20],
    'class_weight': ['balanced', None]       # handle imbalance
}

grid = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    param_grid,
    cv=5,
    scoring='f1',      # use F1 since you have imbalance
    n_jobs=-1,
    verbose=1
)

grid.fit(X_train, y_train)
print("ðŸ”¥ Best Parameters:", grid.best_params_)
print("ðŸŽ¯ Best F1 (CV):", grid.best_score_)
```
