```python
import pandas as pd
df=pd.read_csv(r"E:\Personalized Financial Offer Recommendation System\Learning\decision tree\loan_prediction_datset\loan_data.csv")
```

```python
print(df.columns.tolist())
```

```python
import pandas as pd
from pycaret.classification import setup, get_config

#  Step 1: Clean any weird column names or NaNs
df.columns = df.columns.astype(str).str.strip()
df = df.loc[:, ~df.columns.isna()]  # remove any NaN column headers
df = df.dropna(subset=['loan_status'])  # just to be sure

#  Step 2: Run PyCaret setup
s = setup(
    data=df,
    target='loan_status',
    session_id=42,
    remove_outliers=True,
    outliers_threshold=0.10,
    remove_multicollinearity=True,
    multicollinearity_threshold=0.95,
    preprocess=True,
    train_size=0.80,  # ensures the whole dataset is preprocessed
    verbose=False
)

#  Step 3: Get the preprocessed data
processed_df = get_config('dataset')

#  Step 4: Save or check it
processed_df.to_csv('processed_data.csv', index=False)
print(" Preprocessing complete! File saved as processed_data.csv")
```

```python
#  Step 1: Import and fetch transformed data from PyCaret
from pycaret.classification import get_config
import pandas as pd

# Get transformed train/test splits + target
X_train_transformed = get_config('X_train_transformed')
X_test_transformed = get_config('X_test_transformed')
y_train = get_config('y_train')
y_test = get_config('y_test')

print("Shapes before cleaning:")
print("X_train:", X_train_transformed.shape)
print("X_test:", X_test_transformed.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)

#  Step 2: Reset indexes to prevent misalignment
X_train_transformed = X_train_transformed.reset_index(drop=True)
X_test_transformed = X_test_transformed.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

#  Step 3: Ensure same columns between train and test (safe alignment)
X_test_transformed = X_test_transformed.reindex(columns=X_train_transformed.columns, fill_value=0)

#  Step 4: Combine X and y for both splits
train_final = pd.concat([X_train_transformed, y_train], axis=1)
test_final = pd.concat([X_test_transformed, y_test], axis=1)

#  Step 5: Stack train and test vertically
final_data = pd.concat([train_final, test_final], axis=0).reset_index(drop=True)

#  Step 6: Final verification
print("\n Combined data info:")
print(final_data.info())
print("Any NaNs?", final_data.isnull().sum().sum())
print("Shape:", final_data.shape)

#  Step 7: Save to CSV
final_data.to_csv('model_ready_full_data_clean.csv', index=False)
print("\n Saved final NaN-free model-ready dataset as 'model_ready_full_data_clean.csv' ")
```

```python
from pycaret.classification import get_config
import pandas as pd

#  Step 1: Fetch from current PyCaret experiment
X_train_transformed = get_config('X_train_transformed').reset_index(drop=True)
X_test_transformed = get_config('X_test_transformed').reset_index(drop=True)
y_train = get_config('y_train').reset_index(drop=True)
y_test = get_config('y_test').reset_index(drop=True)

print("Shapes:")
print("X_train:", X_train_transformed.shape)
print("y_train:", y_train.shape)
print("X_test:", X_test_transformed.shape)
print("y_test:", y_test.shape)

#  Step 2: Sanity check  align sizes
if len(X_train_transformed) != len(y_train):
    print(" Mismatch detected in training set sizes! Truncating to shortest length.")
    min_len = min(len(X_train_transformed), len(y_train))
    X_train_transformed = X_train_transformed.iloc[:min_len, :].reset_index(drop=True)
    y_train = y_train.iloc[:min_len].reset_index(drop=True)

if len(X_test_transformed) != len(y_test):
    print(" Mismatch detected in test set sizes! Truncating to shortest length.")
    min_len = min(len(X_test_transformed), len(y_test))
    X_test_transformed = X_test_transformed.iloc[:min_len, :].reset_index(drop=True)
    y_test = y_test.iloc[:min_len].reset_index(drop=True)

#  Step 3: Ensure identical columns
X_test_transformed = X_test_transformed.reindex(columns=X_train_transformed.columns, fill_value=0)

#  Step 4: Combine splits properly
train_final = pd.concat([X_train_transformed, y_train], axis=1)
test_final = pd.concat([X_test_transformed, y_test], axis=1)

#  Step 5: Stack vertically
final_data = pd.concat([train_final, test_final], axis=0).reset_index(drop=True)

#  Step 6: Verify
print("\n Combined Data Info:")
print(final_data.info())
print("Any NaNs?", final_data.isnull().sum().sum())
print("Shape:", final_data.shape)

final_data.to_csv('model_ready_full_data_clean.csv', index=False)
print("\n Saved as model_ready_full_data_clean.csv successfully!")
```
