```python
import pandas as pd
import joblib

# Load pipeline again (just like Streamlit will do)
pipeline = joblib.load("loan_approval_pipeline.pkl")

# üßæ Simulated User Input (like Streamlit form values)
user_input = {
    "person_age": 32,
    "person_gender": "female",
    "person_education": "Master",
    "person_income": 85000,
    "person_emp_exp": 8,
    "person_home_ownership": "MORTGAGE",
    "loan_amnt": 18000,
    "loan_intent": "DEBTCONSOLIDATION",
    "loan_int_rate": 10.2,
    "loan_percent_income": 0.21,
    "cb_person_cred_hist_length": 6,
    "credit_score": 720,
    "previous_loan_defaults_on_file": "No"
}

# Convert to DataFrame (since model expects tabular input)
input_df = pd.DataFrame([user_input])

# Make prediction
pred = pipeline.predict(input_df)[0]
prob = pipeline.predict_proba(input_df)[0][1]

# Display result like Streamlit would
if pred == 1:
    print(f"‚úÖ Loan Approved ‚Äî Confidence: {prob*100:.2f}%")
else:
    print(f"‚ùå Loan Rejected ‚Äî Confidence: {(1 - prob)*100:.2f}%")
```
